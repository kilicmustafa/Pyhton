```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
%matplotlib inline
```

```python
x = np.arange(1,6)
```

```python
x
```

```python
y = np.arange(2,11 ,2)
```

```python
y
```

```python
plt.plot(x ,y ,"red")
plt.show()
```

```python
plt.plot(x ,y)
plt.xlabel("X values")
plt.ylabel("y values")
plt.title("X vs y Graph")
plt.show()
```

```python
plt.subplot(2 ,2 ,1)
plt.plot(x, y **3  ,"pink")

plt.subplot(2 ,2 ,2)
plt.plot(x**2 ,y , "yellow")


plt.subplot(2 , 2, 3)
plt.plot(x*y ,y ,"green")


plt.subplot(2 ,2 ,4)
plt.plot(x ,y *1.5 , "purple")
plt.show()
```

```python
fig = plt.figure()
axes1 = fig.add_axes([.1 , .2 , .3 ,.4])

axes1.plot(x, y ,"blue")
axes1.set_xlabel("X Values")
axes1.set_ylabel("Y Values")
axes1.set_title("X vs Y GRAPH")

plt.show()
```

```python
fig = plt.figure()
axes1 =  fig.add_axes([.3 ,.1 ,.4 ,.5])

axes1.plot(x ,y , "aqua")
axes1.set_xlabel("X values")
axes1.set_ylabel("Y values")
axes1.set_title("X vs Y Graph")

plt.show()
```

```python
fig = plt.figure()
axes1 = fig.add_axes([.1 ,.1 ,.4 ,.4])
axes2 = fig.add_axes([.5 ,.5 ,.4 ,.4])


axes1.plot(x, y*3 ,"red" )
axes1.set_xlabel("Xlabel")
axes1.set_ylabel("Ylabel")
axes1.set_title("Axes 1")


axes2.plot(x**4 ,y , "purple")
axes2.set_xlabel("x label")
axes2.set_ylabel("y label")
axes2.set_title("Axes 2 ")

plt.show()
```

```python
fig ,axes = plt.subplots(nrows = 2 ,ncols = 2)
plt.tight_layout()
```

```python
fig ,axes = plt.subplots(nrows = 2 ,ncols = 1)
plt.tight_layout()
```

```python
fig , axes = plt.subplots(nrows = 2  , ncols = 1)
plt.tight_layout()

for ax in axes:
    ax.plot(x,y)
    
plt.show()
```

```python
fig ,axes = plt.subplots( nrows = 2  ,ncols = 1)
plt.tight_layout()
axes[0].plot(x ,y**3 ,"pink")

axes[1].plot(x**2 ,y ,"aqua")

plt.show()
```

```python
fig = plt.figure(figsize=(6,6))
ax = fig.add_axes([.3 , .3 ,.6 ,.6 ])
ax.plot(x,y**4 ,"pink")

plt.show()
```

```python
fig ,axes  = plt.subplots(nrows = 2 ,ncols=2)
plt.tight_layout()

axes[0,0].plot(x,y**2 ,"red")
axes[0,0].set_title("axes 1")

axes[0,1].plot(x ,y*y ,"pink")
axes[0,1].set_title("axes 2")

axes[1,0].plot(x*x ,y ,"aqua")
axes[1,0].set_title("axes 3")

axes[1,1].plot(x*y,y ,"blue" ,label ="axes 4")
axes[1,1].set_title("axes 4")
axes[1 ,1].legend()

plt.show()

```

```python
fig ,axes = plt.subplots(nrows = 1 ,ncols =2)
axes[0].plot(x ,x**2 ,color = "aqua" ,label = "Aque Value")
axes[0].plot(x**2 ,y ,color = "red" ,label = "Red value")
axes[0].plot(x**3 ,y ,color  ="pink" ,label ="Pink value")
axes[0].plot(y**3 ,x*2 ,color="blue" ,label="Blue value")
axes[0 ].legend()


axes[1].plot(x ,x*1.5 ,color  ="orange" ,label ="Orange Value")
axes[1].plot(y*x ,x*y , color ="yellow" ,label ="Yellow value")
axes[1].plot(x*x ,x/y ,color ="green" ,label="Green Value")
axes[1].legend()

plt.show()
```

```python
fig = plt.figure()
ax = fig.add_axes([.1 ,.2 , 1 , 1])

ax.plot(x ,x**2 ,color="red" ,lw = 7 ,ls ="-." ,marker="o",ms=20 ,mfc="black" ,mec="yellow" ,mew =10)
ax.set_ylim(0,50)
ax.set_xlim(0,10)
plt.show()
```
