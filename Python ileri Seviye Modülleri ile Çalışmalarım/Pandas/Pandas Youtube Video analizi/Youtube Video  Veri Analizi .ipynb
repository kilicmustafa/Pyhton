##### Her bir videoda kullanılan etiket sayısını hesaplayarak, oluşan değerleri yeni bir sütun olarak ekleyin.

Not: Her bir etiket "|" işaretiyle ayrılmaktadır.

```python
def tags(youtube):
    liste = list()
    for i in range(0,len(youtube.index)):
        tag = youtube["tags"].iloc[i].split("|")
        
        liste.append(len(tag))
            
    return liste
            
youtube["bellaCialo"] = tags(youtube)

#-----------------------


def applytag(tags):
    liste = tags.split("|")
    return len(liste)

youtube["newcond"] = youtube["tags"].apply(applytag)
            
            
```
